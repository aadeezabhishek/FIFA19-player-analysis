```python
Barcelona = team.get_group('FC Barcelona')
Barcelona.head()
Barcelona['Wage']= Barcelona['Wage'].str.replace('€','').str.replace('K','')
Barcelona['Wage'] = Barcelona['Wage'].astype(int)

plt.style.use('ggplot')
bar = Barcelona.sort_values('Wage',ascending=False).head(10)
ax = bar.plot(kind='barh',x='Name',y='Wage',color='#61D199',
             title='Highest earning player in FC Barcelona')
ax.set_xlabel('Weekly Earnings')

Barcelona[['Name', 'Value', 'Wage', 'Nationality']].head().style.bar(subset='Wage', color='#61D199', align='zero')
```
